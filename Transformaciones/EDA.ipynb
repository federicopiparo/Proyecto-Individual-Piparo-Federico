# EDA

El Análisis Exploratorio de Datos (Exploratory Data Analysis o EDA) es un paso crucial en el análisis de datos que implica examinar y visualizar un conjunto de datos antes de hacer cualquier modelado o análisis profundo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_parquet ("C:\\Users\\fede\\Desktop\\LABs 1\\Proyecto individual - Federico Piparo\\Transformaciones\\transformados.parquet")
```

### analisis basicos:

```python
# Chequeo de nulos
nulos_cantidad = df.isnull().sum()
nulos_porcentaje = (df.isnull().sum() / df.shape[0]) * 100

# Crear un DataFrame con los resultados
nulos_df = pd.DataFrame([nulos_cantidad, nulos_porcentaje], index=['Cantidad de Nulos', 'Porcentaje de Nulos'])
pd.set_option('display.max_columns', None)
```

```python
nulos_df
```

```python
# Convertir columnas con diccionarios a cadenas de texto para evitar errores de tipo unhashable
df_limpio = df.applymap(lambda x: str(x) if isinstance(x, dict) else x)

# Valores duplicados
print("\nChequeo de valores duplicados:")
print(f"Cantidad de valores duplicados: {df_limpio.duplicated().sum()}")

```

```python
# Valores faltantes
print("\nChequeo de valores faltantes:")
missing_data = df.isnull().sum().reset_index()
missing_data.columns = ['Column', 'Missing Values']
missing_data['Percentage'] = (missing_data['Missing Values'] / df.shape[0]) * 100
print(missing_data)
```

```python
# Valores faltantes
print("\nChequeo de valores faltantes:")
missing_data = df.isnull().sum().reset_index()
missing_data.columns = ['Column', 'Missing Values']
missing_data['Percentage'] = (missing_data['Missing Values'] / df.shape[0]) * 100
print(missing_data)
```

```python
# Tipos de datos
print("\nTipos de datos:")
print(df.dtypes)
```

```python
# Ajustar las opciones de visualización de pandas
pd.set_option('display.max_columns', None)

# Análisis estadístico de datos
estadisticas_descriptivas = df.describe()

print("Análisis estadístico de datos:")
print(estadisticas_descriptivas)
```

```python
# Filtrar solo las columnas numéricas
df_limpio = df.select_dtypes(include=[np.number])

# Análisis de correlaciones
print("\nAnálisis de correlaciones:")
corr_matrix = df_limpio.corr()
print(corr_matrix)

# Visualización del mapa de calor de las correlaciones
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de calor de correlaciones')
plt.show()
```
